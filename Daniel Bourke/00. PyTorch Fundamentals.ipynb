```python
import torch
print(torch.__version__)
```

```python
# Scalar
scalar = torch.tensor(7)
print("Scaler : ", scalar)
print("Scaler dtype : ", scalar.dtype)
print("Scaler shape : ", scalar.shape)
print("Scaler ndim : ", scalar.ndim)
print("Scaler item() : ", scalar.item())
print("Scaler item : ", scalar.item)
```

```python
# Vector
vector = torch.tensor([7, 7])
print("Vector : ", vector)
print("Vector dtype : ", vector.dtype)
print("Vector shape : ", vector.shape)
print("Vector ndim : ", vector.ndim)
#print(vector.item())
print("Vector item : ", vector.item)
```

```python
# Matrix
Matrix = torch.tensor([[11, 22],
                       [33, 44]])

print("Maxtrix : ", Matrix)
print("Maxtrix dtype : ", Matrix.dtype)
print("Maxtrix shape : ", Matrix.shape)
print("Maxtrix ndim : ", Matrix.ndim)
#print(vector.item())
print("Maxtrix item : ", Matrix.item)
```

```python
# Tensor
Tensor = torch.tensor([[[11]],[[11]],[[11]]])
print("Tensor : ", Tensor)
print("Tensor dtype : ", Tensor.dtype)
print("Tensor shape : ", Tensor.shape)
print("Tensor ndim : ", Tensor.ndim)
#print(vector.item())
print("Tensor item : ", Tensor.item)
```
